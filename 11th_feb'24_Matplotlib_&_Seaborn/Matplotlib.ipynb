# plotting 2 graph in one figure horizontaly (side by side)

```python
plt.figure() # to create a empty figure or a page 
plt.subplot(1,2,1) # 1 here is to create a page, 2 is to make 2 section in 1page, 1 is to draw graph in 1st section
plt.plot(xls,np.sin(xls),'.-')

plt.subplot(1,2,2)
plt.plot(xls,np.cos(xls),'.-')
plt.show() # to show the graph
```

# plotting 4 graph in 2 figure

```python
plt.figure()  
plt.subplot(2,2,1) 
plt.plot(xls,np.sin(xls),'.-')

plt.subplot(2,2,2)
plt.plot(xls,np.cos(xls),'.-')

plt.subplot(2,2,3)
plt.plot(xls,np.tan(xls),'.-')

plt.subplot(2,2,4)
plt.plot(xls,np.cos(xls+1),'.-')

plt.show() 
```

# creating the plot in vertical mannner up and down 

```python
plt.figure() # to create a empty figure or a page 
plt.subplot(2,1,1) # 1 here is to create a page, 2 is to make 2 section in 1page, 1 is to draw graph in 1st section
plt.plot(xls,np.sin(xls),'.-')

plt.subplot(2,1,2)
plt.plot(xls,np.cos(xls),'.-')
plt.show() # to show the graph
```

```python
plt.plot(xls,np.sin(xls-0),color = 'blue')
plt.axis('equal')
plt.grid()
plt.show()
```

# adding title:

```python
plt.plot(xls,np.sin(xls-0),'.-',color = 'blue')
plt.axis('equal')
plt.grid()
plt.title('This is sin plot before real dataset execution to understand basic syntax')
plt.show()
```

# labelling X and Y -axis:

```python
plt.plot(xls,np.sin(xls-0),'.-',color = 'blue')
plt.axis('equal')
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('This is sin plot before real dataset execution to understand basic syntax')
plt.show()
```

# add label for the plot lines

```python
plt.plot(xls,np.sin(xls-0),color = 'blue',label='Sinx')
plt.plot(xls,np.cos(xls-0),color = 'red',label= 'Cosx')
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('This is sin plot before real dataset execution to understand basic syntax')
plt.show()
```

# Histogram:

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
# changing the bin size:
plt.hist(data, bins = 20)
plt.show()
```

```python
# changing the histpgram type and color:
#histtype; supported values are 'bar', 'barstacked', 'step', 'stepfilled'
```

```python
plt.hist(data, bins = 20,histtype='step' , color ='red')
plt.show()
```

```python
plt.hist(data, bins = 20, histtype ='bar', color ='green')
plt.show()
```

```python
plt.hist(data, bins = 20, histtype ='stepfilled', color ='red')
plt.show()
```

```python
# to make color transparent: 
# using alpha
```

```python
plt.hist(data, bins = 20,alpha=0.7, histtype ='barstacked', color ='red')
plt.show()
```

```python
plt.hist(data, bins = 20,alpha=0.2, histtype ='stepfilled', color ='red')
plt.show()
```

# Scatter plot:

```python
plt.scatter(x,y,c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(6,2)) # figsize have widh and height in inches
plt.scatter(x,y,c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("with figsize")
plt.show()
```

# Bar plot:

```python
x = ['a','b','c','d'] # categorial data 
y = [3,4,1,6]
```

```python
plt.figure(figsize= (6,3))
plt.bar(x,y, color = 'g')
plt.title("This is my bar plot")
plt.show()
```

```python
plt.figure(figsize= (6,3))
plt.bar(x,y, color = '#ded00b')
plt.title("This is my bar plot")
plt.show()
```

```python
# horizontal bar plot
plt.figure(figsize= (6,3))
plt.barh(x,y)
plt.title("This is my bar plot")
plt.show()
# it is ideal to use the default blue color
```

# 3-d plot:

```python
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c='g',alpha =0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis') # can't label for z as it is not available it will show error
plt.title("3-D Plot")
plt.show()
```

# Working with dataset:

```python
data = pd.read_csv("Bank_churn.csv")
```

```python
data.head()
```

```python
data.plot.scatter(x='Age',y='Balance')
```

```python
# insights from above plot:
#maximum concentration >> 25 -50
# maximum balance >> 40
# min or zero balance >> in all the age group
```

```python
# age distribution
plt.title("Age Distribution")
plt.hist(data['Age'])
plt.xlabel("Age")
plt.ylabel("Number of person")
plt.show()
# Balance Distribution
plt.title("Balance Distribution")
plt.hist(data['Balance'])
plt.xlabel("Balance")
plt.ylabel("Number of person")
plt.show()
```

# Piechart

```python
langs = ['C','Python','Java']
students = [20,100,40]
plt.pie(students,labels=langs,autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
my_explode = (0.0,0.1,0.1)

langs = ['C','Python','Java']
students = [20,100,40]
plt.pie(students,labels=langs,autopct = '%1.1f%%',explode=my_explode,shadow=True)
plt.title("Pie chart")
plt.show()
```
