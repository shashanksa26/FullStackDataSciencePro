```python
import seaborn as sns
```

```python
sns.get_dataset_names() # built-in dataset in seaborn
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

# Histogram:

```python
sns.histplot(tips['total_bill'])
```

# Distrogram:

```python
sns.distplot(tips['total_bill'],bins =50)
```

```python
# the curve around the plot is kde = kernal density estimeter
#density is probability density : it is the probability of finding data between x-axis values
```

```python
# addig bins and removing the kde : without kde it is just a hsitogram
sns.distplot(tips['total_bill'],bins=30, kde =False)
```

# scatter plot:

```python
sns.scatterplot(x=tips.total_bill , y=tips.tip)
```

```python
# checking the relationship of totalbill and tip with smoker
sns.scatterplot(x=tips.total_bill , y=tips.tip,data =tips, hue='smoker')
```

```python
# instead of different color we can differentiate the 3rd parameter by different style
sns.scatterplot(x=tips.total_bill , y=tips.tip,data =tips, style='size')
```

```python
# instead of different color we can differentiate the 3rd parameter by different style
sns.scatterplot(x=tips.total_bill , y=tips.tip,data =tips, style='smoker')
```

```python
sns.scatterplot(x=tips.total_bill , y=tips.tip,data =tips, style='smoker', hue = 'sex')
```

```python
# insight 
# it is linear relationship as total bill is increasing and the tip is increasing
# maximum tip is paid by maximum total_bill
```

# relplot:

```python
sns.relplot(x=tips.total_bill , y=tips.tip,data =tips, hue='smoker')
```

```python
sns.relplot(x=tips.total_bill , y=tips.tip,data =tips, hue='size')
```

```python
# instead of different color we can differentiate the 3rd parameter by different style
sns.relplot(x=tips.total_bill , y=tips.tip,data =tips, style='size')
```

```python
# we need both style and hue for 3rd and 4th parameter
sns.relplot(x=tips.total_bill , y=tips.tip,data =tips, style='smoker', hue = 'time')
```

# catplot:

```python
# Maximum total_bill generated in which day
sns.catplot(x='day',y='total_bill', data=tips)
```

# barplot:

```python
# it is givng the count of total number of visitors
sns.barplot(x='day',y='total_bill', data=tips)
```

# jointplot:

```python
# it gives distribution and the data together : the distribution is given as bar opposite to it's axis
sns.jointplot(x ='total_bill', y = 'tip' ,data=tips) 
```

```python
# adding regression line in the jointplot
sns.jointplot(x ='total_bill', y = 'tip' ,data=tips,kind ='reg') 
```

```python
# adding hexagoan in the jointplot
sns.jointplot(x ='total_bill', y = 'tip' ,data=tips,kind ='hex') 
```

```python
# adding kde in the jointplot
sns.jointplot(x ='total_bill', y = 'tip' ,data=tips,kind ='kde') #contour lines > meterological dept., geo phys
```

# Heat map

```python
# correlation: to understand the relationship between two different data and their dependency
# it is done on the numerical data only
```

```python
tips
```

```python
# getting the numrical columns from the tips dataset as it is having string too
tips_numerical = tips[['total_bill','tip','size']] 
```

```python
# to find correlation:
tips_numerical.corr() 
```

```python
# heatmap plotting
sns.heatmap(tips_numerical.corr(),annot=True)
```

```python
# to change the color mapping 
sns.heatmap(tips_numerical.corr(),annot=True,cmap='rainbow')
```

 # lmplot:
